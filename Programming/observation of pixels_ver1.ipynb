```python
# observation from the ground

import numpy as np
import matplotlib.pyplot as plt

v_car=5.611 # 20km/h on average in hk
v_eye=16 # Hz
alpha_karry=1.00
v_reload=alpha_karry*v_eye

pie=math.pi
r_g=2000
rho_car=0.001023
delta_d_car=v_car/v_reload
L_car=4.71769
C_eye=40960000
alpha_c1=1.00
C1=C_eye*alpha_c1
alpha_c2=0.90
C2=C_eye*alpha_c2
alpha_c3=0.60
C3=C_eye*alpha_c3

fig,(ax1,ax2,ax3)=plt.subplots(3,figsize=(15,5))
fig.suptitle('Observation from the Ground')

XLEFT=0
XRIGHT=1
YDOWN=4.4e13
YUP=4.92e13
datakarry=np.arange(XLEFT+0.01,XRIGHT,0.01)
# plt.figure(figsize=(10,10))
# plt.xlim((XLEFT,XRIGHT))
# plt.ylim((YDOWN,YUP))
# xt=np.arange(XLEFT,XRIGHT,(XRIGHT-XLEFT)/10)
# yt=np.arange(YDOWN,YUP,(YUP-YDOWN)/10)
# plt.xticks(xt)
# plt.yticks(yt)

# print(datakarry)

def alpha_k(x):
    if x>1:
        return 1
    elif x>= 1-np.log(2) and x<=1:
        return (2-np.exp(1-x))
    elif x>0 and x< 1-np.log(2):
        return 0

datayc1=[0 for i in range(len(datakarry))] # when C=100%*C_eye
datayc2=[0 for i in range(len(datakarry))] # when C=90%*C_eye
datayc3=[0 for i in range(len(datakarry))] # when C=60%*C_eye
dataoutkarry=[0 for i in range(len(datakarry))]
for i in range(len(datakarry)):
    datayc1[i]=(4*pie*(r_g**4)*rho_car*(v_car/(v_eye*datakarry[i]))*L_car*C1)/(3*((r_g-1)**2))
    datayc2[i]=(4*pie*(r_g**4)*rho_car*(v_car/(v_eye*datakarry[i]))*L_car*C2)/(3*((r_g-1)**2))
    datayc3[i]=(4*pie*(r_g**4)*rho_car*(v_car/(v_eye*datakarry[i]))*L_car*C3)/(3*((r_g-1)**2))
    dataoutkarry[i]=alpha_k(datakarry[i])
print(dataoutkarry)
ax1 = plt.subplot(131)
ax1.plot(datakarry,datayc1)
ax1.set_xlabel('alpha_karry')
ax1.set_ylabel('number of pixels')
# ax1.legend(['alpha_c=100%'])

# ax2 = plt.subplot(132)
# ax2.plot(datakarry,datayc2)
# ax2.set_xlabel('alpha_karry')
# ax2.set_ylabel('number of pixels')
# ax2.legend(['alpha_c=90%'])

ax1.plot(datakarry,datayc2)
ax1.plot(datakarry,datayc3)
ax1.legend(['alpha_c=100%','alpha_c=90%','alpha_c=60%','karry_index'])

ax2=plt.subplot(132)
ax2.plot(datakarry,dataoutkarry)
ax2.set_xlabel('velocity portion')
ax2.set_ylabel('alpha_karry')
ax2.legend(['karry index'])


annotated_points=[0.6,0.9,1.0]
annotated_outputs=[alpha_clarity(i) for i in annotated_points]

ax3 = plt.subplot(133)
ax3.plot(dataport,output,c='r')
ax3.scatter(annotated_points,annotated_outputs,c='r')
ax3.set_xlabel('pixels compared with the human eye standard')
ax3.set_ylabel('alpha_clarity')
ax3.legend(['alpha_c','points in the graphs'])

fig.tight_layout()
plt.subplots_adjust(wspace = 0.3)
plt.show()
```
